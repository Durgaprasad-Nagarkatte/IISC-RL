```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="darkgrid")
```

```python
np.random.seed(0)
mu_rewards = np.random.randn(10)*10
mu_rewards = [np.round(reward, decimals=2) for reward in mu_rewards]
print(mu_rewards)
```

```python
x = np.linspace(0, 10, 10)
plt.plot(x, mu_rewards, 'o', color='black');
```

```python
#Function for generating the reward with the given mu and variance
def generate_reward(index):
    return np.round(np.random.normal(mu_rewards[index], 1), decimals=2)
```

```python
#Checking the working on plot
rewards = []
for i in range(1000):
    reward = generate_reward(0)
    rewards.append(reward)
x = np.linspace(-(mu_rewards[0] + 3) , (mu_rewards[0] + 3), 1000) 
plt.scatter(x, rewards, alpha=0.5)
```

```python
def generate_action():
    action = np.random.randint(low=0, high=9)
    return action
```

```python
actions = []
for i in range(100):
    actions.append(generate_action())
print(actions)
```

```python
def simulation_for_one_instant_greedy():
    Qt = []
    Q = [0]*10
    N = [0]*10
    action = generate_action()
    for i in range(1000):
        reward = generate_reward(action)
        N[action] += 1
        Q[action] = Q[action] + (1/N[action])*(reward-Q[action])
        Qt.append(Q[action])
    return Q, N, Qt
```

```python
Q, N, Qt = simulation_for_one_instant_greedy()
print(Q)
print(N)
```

```python
x = np.linspace(0, 1000, 1000)
plt.plot(x, Qt)
```

```python
def simulation_for_one_greedy_epsilon(epsilon):
    Qt = []
    Q = [0]*10
    N = [0]*10
    not_explored_actions = [i for i in range(10)]
    for i in range(1000):
        prob = np.random.rand()
        if(len(not_explored_actions) > 0):
            if prob <= epsilon:
                action = np.random.choice(not_explored_actions)
                index = not_explored_actions.index(action)
                not_explored_actions.pop(index)
            else:
                action = np.argmax(Q)
        else:
            action
        reward = generate_reward(action)
        N[action] += 1
        Q[action] = Q[action] + (1/N[action])*(reward-Q[action])
        Qt.append(Q[action])
    return Q, N, Qt
```

```python
epsilon = 0.5
Q, N, Qt = simulation_for_one_greedy_epsilon(epsilon)
print(Q)
print(N)
```

```python
x = np.linspace(0, 1000, 2000)
plt.plot(x, Qt)
```

```python
def generate_multiple_simulations_greedy():
    Qt_multiple = []
    for i in range(2000):
        _, _, Qt = simulation_for_one_instant_greedy()
        Qt_multiple.append(Qt)
    return Qt_multiple
```

```python
Qt_multiple = generate_multiple_simulations_greedy()
Qt_multiple = np.array(Qt_multiple)
Qt_multiple.shape
```

```python
newQt = Qt_multiple.mean(axis=0)
print("The shape of newQt", newQt.shape)
```

```python
x = np.linspace(0, 1000, 1000)
plt.plot(x, newQt)
```

```python
def generate_multiple_simulations_greedy_epsilon(epsilon):
    Qt_multiple = []
    for i in range(2000):
        _, _, Qt = simulation_for_one_greedy_epsilon(epsilon)
        Qt_multiple.append(Qt)
    return Qt_multiple
```

```python
epsilon = 0.5
Qt_multiple = generate_multiple_simulations_greedy_epsilon(epsilon)
Qt_multiple = np.array(Qt_multiple)
Qt_multiple.shape
```

```python
newQt = Qt_multiple.mean(axis=0)
print("The shape of newQt", newQt.shape)
```

```python
x = np.linspace(0, 1000, 1000)
plt.plot(x, newQt)
```
